## Maximum Depth of Binary Tree

Given the root of a binary tree, returns its *maximum depth*.  

A binary tree's **maximum depth** is the number of nodes along the longest path from the root node down to the farthest leaf node. 

#### Example 1:  



```
Input: root = [3, 9, 20, null, null, 15, 17]
Output: 3
```

Given a Binary Tree, **Level Order Traversal** technique is a method to traverse a Tree such that all nodes present in the same level are traversed completely before traversing the next level.

### How does Level Order Traversal work?

The main idea of level order traversal is to traverse all the nodes of a lower level before moving to ant of the nodes of a higher level. This can be done in any of the following ways:
- Using Stack (Recursion)
- Using Queue (Iteration)

- [Approach 2] Using Queue (Iterative) - O(n) time and O(n) Space
