# Binary Addition

Adding in binary is significantly simpler than adding in decimal when you keep in mind the following tips or guidelines. Let’s take a look at the rules of binary addition.

### How to Do Binary Addition

Addition of binary numbers includes two cases:

- Binary addition with regrouping
- Binary addition without regrouping

Let us understand binary addition step by step.

## Problem Add Binary

Given two binary strings a and b, return their sum as a binary string.

Example 1:
Input: a = "11", b = "1"
Output: "100"

Example 2:
Input: a = "1010", b = "1011"
Output: "10101"

```python
def addBinary(a: str, b: str):
    # Retrieve the longest binary number, and adjust the lenght of the two binary to be equal.
    longest_binary = len(a) + 1 if len(a) >= len(b) else len(b) + 1
    while len(a) < longest_binary or len(b) < longest_binary:
        if len(a) < longest_binary:
            a = "0" + a
        elif len(b) < longest_binary:
            b = "0" + b
    sum_rules_dict = {"00": "0", "01": "1", "10": "1", "11": "0"}    
    carry_over = ''.join(["0" for i in range(longest_binary)])
    result_binary_sum = ""
    for i in range(len(a))[::-1]:
        # Sum with the carry over.
        merged_values = carry_over[i] + a[i]
        column_sum = ""
        if merged_values == "11":
            carry_over_list = [el for el in carry_over]
            carry_over_list[i - 1] = "1"
            carry_over = ''.join(carry_over_list)
        column_sum = sum_rules_dict[merged_values]
        # Sum the result with the second binary value (b)
        merged_values = column_sum + b[i]
        if merged_values == "11":
            carry_over_list = [el for el in carry_over]
            carry_over_list[i - 1] = "1"
            carry_over = ''.join(carry_over_list)
        column_sum = sum_rules_dict[merged_values]
        result_binary_sum = column_sum + result_binary_sum  

    if result_binary_sum[0] == "0" : result_binary_sum = result_binary_sum[1:]
    
    return result_binary_sum

a = "1011"
b = "111"
print(addBinary(a, b))
```
