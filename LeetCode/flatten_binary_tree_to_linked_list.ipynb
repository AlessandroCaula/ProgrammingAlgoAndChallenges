# Flatten Binary Tree to Linked List

Given the `root` of a binary tree, flatten the tree into a "linked list":

- The "linked list" should use the same `TreeNode` class where the `right` child pointer points to the next node in the list and the `left` child pointer is always `null`.

- The "linked list" should be in the same order as a **pre-order traversal** of the binary tree.

### Example:



```
Input: root = [1,2,5,3,4,null,6]
Output: [1,null,2,null,3,null,4,null,5,null,6]
```

## Solution

We can solve this problem *recursively* or *iteratively*. The recursive solution is simpler to understand and that's what we'll cover today.

For each node, we will:

1. Recursively flatten the `left` subtree.

2. Recursively flatten the `right` subtree.

3. Set the node.left to `null`.

4. Adjust node.right to match out required format: node  Flattened left subtree  Flattened right subtree

The time complexity of this solution is `O(n)` where `n` is the number of nodes in the tree.

````python
def flatten(root):
    # Base case: empty tree or leaf node
    if not root:
        return

    # Store right and left subtrees
    right = root.right
    left = root.left

    # Set left to None (requirement of the problem)
    root.left = None

    # Recursively flatten left subtree
    flatten(left)
    # Recursively flatten right subtree
    flatten(right)

    # Connect flattened left subtree to root
    root.right = left

    # Find the end of left subtree
    current = root
    while current.right:
        current = current.right

    # Connect flattened right subtree
    current.right = right
````
