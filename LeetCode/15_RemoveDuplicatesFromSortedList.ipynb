Given the head of a sorted linked list, delete all duplicates such that each element appears only once. Return the *linked list ***sorted*** as well*.

Exaple 1:



```
Input: head = [1,1,2]
Output: [1,2]
```

Example 2:



```
Input: head = [1,1,2,3,3]
Output: [1,2,3]
```

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

    def __repr__(self):
        return self.val

class LinkedList:
    def __init__(self, nodes=None):
        self.head = None
        if nodes is not None:
            node = Node(val=nodes.pop(0))
            self.head = node
            for elem in nodes:
                node.next = Node(val=elem)
                node = node.next
    
    def __repr__(self) -> str:
        node = self.head
        string_llist = str(node.val) + "->"
        while node is not None:
            if self.head == node:
                node = node.next
                continue
            string_llist += str(node.val) + "->"
            node = node.next
        string_llist += "None"
        return string_llist
````

````python
def deleteDuplicates(node):
    head = node
    while node is not None:
        next_node = node.next
        while next_node is not None and next_node.val == node.val:
            next_node = next_node.next
        node.next = next_node
        node = node.next
    return head

llist = LinkedList([1, 1, 1, 1, 1, 2, 3, 3])
removed_double = deleteDuplicates(llist.head)
while removed_double is not None:
    print(removed_double.val)
    removed_double = removed_double.next
````
