# Valid Sudoku

You are given a `9 x 9` Sudoku board `board`. A Sudoku board is valid if the following rules are followed.

1. Each row must contain the digits `1-9` without duplicates
2. Each column must contain the digits `1-9` without duplicates.
3. Each of the nine digits `3 x 3` sub-boxes of the grid must contain the digits `1-9` without duplicates. 

Return `True` if the Sudoku board is valid, otherwise return `False`.

__Note__: A board does not need to be full or be solvable to be valid. 

#### Example 1:



```py
Input: board = 
[["1","2",".",".","3",".",".",".","."],
 ["4",".",".","5",".",".",".",".","."],
 [".","9","8",".",".",".",".",".","3"],
 ["5",".",".",".","6",".",".",".","4"],
 [".",".",".","8",".","3",".",".","5"],
 ["7",".",".",".","2",".",".",".","6"],
 [".",".",".",".",".",".","2",".","."],
 [".",".",".","4","1","9",".",".","8"],
 [".",".",".",".","8",".",".","7","9"]]

Output: true
```

Explanation: there are two 1\'s in the top-left 3x3 sub-box

`````python
def valid_sudoku(board) -> bool:
    
    # First check the first two constraints. 
    # Each row must contain the digits 1-9 without duplicates
    # Each column must contain the digit 1-9 without duplicates
    # --- Checking for duplicates in the rows
    for row in board:
        row_collection = []
        for el in row:
            if el != "." and el in row_collection:
                return False
            else:
                row_collection.append(el)
    # --- Checking for duplicates in the columns
    for col in range(len(board[0])):
        col_collection = []
        for row in range(len(board)):
            el = board[row][col]
            if el != "." and el in col_collection:
                return False
            else:
                col_collection.append(el)
    
    # Compute the number of 3x3 sub-box are present. 
    rows_by_3 = len(board) / 3
    cols_by_3 = len(board[0]) / 3


    return True

board = [["1","2",".",".","3",".",".",".","."],
         ["4",".",".","5",".",".",".",".","."],
         [".","9","8",".",".",".",".",".","3"],
         ["5",".",".",".","6",".",".",".","4"],
         [".",".",".","8",".","3",".",".","5"],
         ["7",".",".",".","2",".",".",".","6"],
         [".",".",".",".",".",".","2",".","."],
         [".",".",".","4","1","9",".",".","8"],
         [".",".",".",".","8",".",".","7","9"]]


print(valid_sudoku(board))
`````
