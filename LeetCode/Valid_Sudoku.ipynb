# Valid Sudoku

Determine if a `9 x 9` Sudoku board is valid. Only the filled cells need to be validated according to the following rules:

1. Each row must contain the digit `1 - 9` without repetition.

2. Each column must contain the digits `1 - 9` without repetition.

3. Each of the nine `3 x 3` sub-boxes of the grid must contain the digits `1 - 9` without repetition.

__Note:__

A Sudoku board (partially filled) could be valid but is not necessarily solvable. 

Only the filled cells need to be validated according to the mentioned rules.

### Example 1:



```
Input: board = 
[["5","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]
Output: true
```

### Example 2:

```
Input: board = 
[["8","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]
Output: false
Explanation: Same as Example 1, except with the 5 in the top left corner being modified to 8. Since there are two 8's in the top left 3x3 sub-box, it is invalid.
```

````python
def valid_sudoku(board):
    # -- Check firs condition: Each row must contain the digits 1-9 without repetition.
    for row in board:
        present_numbers = []
        for el in row:
            if el in present_numbers:
                return False
            elif el != ".":
                present_numbers.append(el)

    # -- Check second condition: Each column must contain the digits 1-9 without repetition.
    for col in range(len(board[0])):
        present_numbers = []
        for row in range(len(board)):
            el = board[row][col]
            if el in present_numbers:
                return False
            elif el != ".":
                present_numbers.append(el)
    
    # -- Check third condition: Each of the nine 3 x 3 sub-boxes of the grid must contain the digits 1-9 without repetition.
    row_box = len(board) // 3
    col_box = len(board[0]) // 3
    for r in range(row_box):
        r_start = r * 3
        for c in range(col_box):
            c_start = c * 3
            present_numbers = []
            # Loop the 3 x 3 grid 
            for i in range(3):
                curr_r = r_start + i
                for j in range(3):
                    curr_c = c_start + j
                    el = board[curr_r][curr_c]
                    if el in present_numbers:
                        return False
                    elif el != ".":
                        present_numbers.append(el)
                    
    return True



board = [
     ["5","3",".",".","7",".",".",".","."]
    ,["6",".",".","1","9","5",".",".","."]
    ,[".","9","8",".",".",".",".","6","."]
    ,["8",".",".",".","6",".",".",".","3"]
    ,["4",".",".","8",".","3",".",".","1"]
    ,["7",".",".",".","2",".",".",".","6"]
    ,[".","6",".",".",".",".","2","8","."]
    ,[".",".",".","4","1","9",".",".","5"]
    ,[".",".",".",".","8",".",".","7","9"]
]

board1 = [
     ["8","3",".",".","7",".",".",".","."]
    ,["6",".",".","1","9","5",".",".","."]
    ,[".","9","8",".",".",".",".","6","."]
    ,["8",".",".",".","6",".",".",".","3"]
    ,["4",".",".","8",".","3",".",".","1"]
    ,["7",".",".",".","2",".",".",".","6"]
    ,[".","6",".",".",".",".","2","8","."]
    ,[".",".",".","4","1","9",".",".","5"]
    ,[".",".",".",".","8",".",".","7","9"]
]

print(valid_sudoku(board))
print(valid_sudoku(board1))
````
