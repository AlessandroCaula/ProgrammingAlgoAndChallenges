# Bingo Check

**28 - 01 - 2026**

Create a function that takes a 5Ã—5 2D list and returns `True` if it has at least one Bingo, and `False` if it doesn't.

## Examples

```py
bingo_check([
  [45, "x", 31, 74, 87],
  [64, "x", 47, 32, 90],
  [37, "x", 68, 83, 54],
  [67, "x", 98, 39, 44],
  [21, "x", 24, 30, 52]
])
output = True

bingo_check([
  ["x", 43, 31, 74, 87],
  [64, "x", 47, 32, 90],
  [37, 65, "x", 83, 54],
  [67, 98, 39, "x", 44],
  [21, 59, 24, 30, "x"]
]) 
output = True

bingo_check([
  ["x", "x", "x", "x", "x"],
  [64, 12, 47, 32, 90],
  [37, 16, 68, 83, 54],
  [67, 19, 98, 39, 44],
  [21, 75, 24, 30, 52]
]) 
output = True

bingo_check([
  [45, "x", 31, 74, 87],
  [64, 78, 47, "x", 90],
  [37, "x", 68, 83, 54],
  [67, "x", 98, "x", 44],
  [21, "x", 24, 30, 52]
]) 
output = False
```

````python
def bingo_check(grid):

    # Check diagonal
    for i in range(len(grid)):
        if grid[i][i] != "x":
            break
        if i == len(grid) - 1:
            return True

    # Loop through all the rows
    for r in range(len(grid)):
        if grid[r][0] == "x":
            # If the first cell of the row contain an "x" check if all the row has all "x"
            for c in range(len(grid[r])):
                if grid[r][c] != "x":
                    break
                if c == len(grid[r]) - 1:
                    return True

    # Loop through all the cols
    for c in range(len(grid[0])):
        if grid[0][c] == "x":
            # If the first cell of the column contain an "x" check if all the column has all "x"
            for r in range(len(grid)):
                if grid[r][c] != "x":
                    break
                if r == len(grid) - 1:
                    return True

    return False


grid = [
  [45, "x", 31, 74, 87],
  [64, "x", 47, 32, 90],
  [37, "x", 68, 83, 54],
  [67, "x", 98, 39, 44],
  [21, "x", 24, 30, 52]
]
print(bingo_check(grid))

grid = [
  ["x", 43, 31, 74, 87],
  [64, "x", 47, 32, 90],
  [37, 65, "x", 83, 54],
  [67, 98, 39, "x", 44],
  [21, 59, 24, 30, "x"]
]
print(bingo_check(grid))

grid = [
  ["x", "x", "x", "x", "x"],
  [64, 12, 47, 32, 90],
  [37, 16, 68, 83, 54],
  [67, 19, 98, 39, 44],
  [21, 75, 24, 30, 52]
]
print(bingo_check(grid))

grid = [
  [45, "x", 31, 74, 87],
  [64, 78, 47, "x", 90],
  [37, "x", 68, 83, 54],
  [67, "x", 98, "x", 44],
  [21, "x", 24, 30, 52]
]
print(bingo_check(grid))
````
